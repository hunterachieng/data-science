```python
nltk.download('stopwords')
```

```python
import string
string.punctuation
```

```python
useless_words = nltk.corpus.stopwords.words("english") + list(string.punctuation)
# If not enough add your own list of stopwords- i.e emoji
print(useless_words)
```

```python
def build_bag_of_words_features_filtered(words):
    return {
        word: 1 for word in words if not word in useless_words
    }
```

```python
build_bag_of_words_features_filtered(movie_texts)
```

```python
def build_bag_of_words_features_unfiltered(words):
    return {
        word: 0 for word in words if word in useless_words
    }
```

```python
build_bag_of_words_features_unfiltered(movie_texts)
```
